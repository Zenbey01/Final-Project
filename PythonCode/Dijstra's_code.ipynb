```python
# ตรงนี้จะเป็นการหาเส้นทางที่มีความเป็นไปได้ทั้งหมด
def all_paths(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return [path]
    if start not in graph:
        return []
    paths = []
    for node in graph[start]:
        if node not in path:
            new_paths = all_paths(graph, node, end, path)
            for new_path in new_paths:
                paths.append(new_path)
    return paths

def all_paths_between_nodes(graph, node1, node2):
    paths_from_node1_to_node2 = all_paths(graph, node1, node2)
    paths_from_node2_to_node1 = all_paths(graph, node2, node1)

# return paths_from_node1_to_node2 + paths_from_node2_to_node1 + paths_from_node1_to_node1 + paths_from_node2_to_node2
#เพิ่มเส้นทาง (paths) จากโหนด (node) 1 ไปยังโหนด 2 และจากโหนด 2 ไปยังโหนด 1 พร้อมกับเส้นทางจากโหนด 1 ไปยังโหนด 1 และเส้นทางจากโหนด 2 ไปยังโหนด 2 แล้วคืนผลลัพธ์ทั้งหมดนี้เป็นผลลัพธ์ของฟังก์ชัน

    return paths_from_node1_to_node2 + paths_from_node2_to_node1

#เพื่อสร้างรายการ filtered_paths โดยผ่านทุกเส้นทาง path ใน paths และเก็บเฉพาะเส้นทางที่เริ่มต้นด้วยโหนดใน condition_nodes
def filter_paths_by_condition(paths, condition_nodes):
    filtered_paths = [path for path in paths if path[0:1] == condition_nodes]
    return filtered_paths

graph = {
    'S': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'X'],
    'A': ['S', 'B', 'C', 'D', 'E', 'F', 'G', 'X'],
    'B': ['S', 'A', 'C', 'D', 'E', 'F', 'G', 'X'],
    'C': ['S', 'A', 'B', 'D', 'E', 'F', 'G', 'X'],
    'D': ['S', 'A', 'B', 'C', 'E', 'F', 'G', 'X'],
    'E': ['S', 'A', 'B', 'C', 'D', 'F', 'G', 'X'],
    'F': ['S', 'A', 'B', 'C', 'D', 'E', 'G', 'X'],
    'G': ['S', 'A', 'B', 'C', 'D', 'E', 'F', 'X'],
    'X': ['S', 'A', 'B', 'C', 'D', 'E', 'F', 'G']
}

start_node1 = 'S'
end_node1 = 'X'

condition_nodes_start_S = ['S']
condition_nodes_start_X = ['X']

all_possible_paths = all_paths_between_nodes(graph, start_node1, end_node1)
filtered_paths1 = filter_paths_by_condition(all_possible_paths, condition_nodes_start_S)
filtered_paths2 = filter_paths_by_condition(all_possible_paths, condition_nodes_start_X)

final_filtered_paths = filtered_paths1 + filtered_paths2

paths = []
paths2 = []

# print("All possible paths:")
for path in final_filtered_paths:
    if len(path) > 1:
        paths.append(path[:-1] + ['X'])

for path in final_filtered_paths:
    if len(path) > 1:
        paths2.append(path[:-1] + ['S'])

final = paths + paths2
for path in final:
    print(path)
```

```python
import heapq
import csv

# function การทำงานของ ai dijkstra กำหนดตัวแปร total_distances คือตัวแปรที่เก็บระยะทางทั้งหมด กำหนดตัวแปร constructed_paths ไว้เก็บค่า dictionary ผลลัพของเส้นทางทั้งหมด ไม่รวม start และ end
def dijkstra(graph, start):
    total_distances = 0
    constructed_paths = []

# กำหนดให้ i = 0 และ loop ในระยะของ ขนาดจำนวนตัวแปร เช่น [A, B, C] ที่ทั้งหมด 3 ตัว ก็จะ loop 0 1 2 ตามลำดับ
# กำหนดให้ตัวแปร distances คือตัวแปรสำหรับเก็บค่า dictionary โดยการเก็บ node ทั้งหมดไว้ในรูปแบบ key โดยจะเก็บค่า node และกำหนดระยะทางของทั้งหมดเป็น inf ในรูปแบบ [A:inf, B:inf] และ ทำการกำหนด node ให้ node จุดเริ่มต้นมีระยะทาง = 0 ( start = A , [A=0, B:inf] )
# และกำหนดให้ตัวแปร pq คือตัวแปรสำหรับเก็บค่า queue ประเภท priority queue คือการใช้ คิวแบบมีลําดับความสําคัญ โดยจะเก็บค่าแรกคือ ( จุดเริ่มต้น, ระยะทาง 0 )
    for i in range(len(start) - 1):
        distances = {node: float('inf') for node in graph}
        distances[start[i]] = 0
        pq = [(0, start[i])]

        path = {node: None for node in graph}

# pop node ที่ระยะทางที่น้อยที่สุดออกจากคิว โดย current_dist เก็บระยะทางขั้นต่ำปัจจุบัน และ current_node เก็บโหนดที่เกี่ยวข้อง
        while pq:
            current_dist, current_node = heapq.heappop(pq)

            if current_node == start[i + 1]:
                break

# #loop ข้อมูลจาก node ปัจจุบัน โดยเก็บค่า node ที่ neighbor และเก็บค่า ระยะทางที่ weight และคำนวนระยะทางปัจจุบัน โดยนำระยะทางปัจจุบัน + ระยะทางจาก loop
            for neighbor, weight in graph[current_node]:
                distance = current_dist + weight

                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(pq, (distance, neighbor))
                    path[neighbor] = current_node

# กำหนดให้ตัวแปร current เก็บค่า end node ซึงคือ node ถัดไปจาก node start (start[i + 1]) พร้อมเก็บค่า end node ลงใน constructed_path
        current = start[i + 1]
        constructed_path = [current]
# หาก loop ค่าที่ไม่ใช่จุดเริ่มต้น ให้เก็บค่าเส้นทางที่ผ่านลงในตัวแปร current พร้อมทั้งทำการเพิ่มตัวแปร current ลงใน ตัวแปร constructed_path และเพิ่มตัวแปร current ลงใน ตัวแปร constructed_paths
        while current != start[i]:
            current = path[current]
            constructed_path.append(current)
            constructed_paths.append(current)

# ทำการเก็บระยะทางปัจจุบันจากจุดเริ่มจนถึงจุด end ทำการบวกรวมจากระยะทางปัจจุบันลงในตัวแปร total_distances เพื่อเก็บไว้สำหรับบอกว่าระยะทางรวมเท่าไหร่
        total_distance = distances[start[i + 1]]
        total_distances += total_distance

# Remove used edges from the graph
# logic สำหรับลบเส้นทางที่เราผ่านไปแล้วออกจาก graph เพื่อไม่ให้เกิดการเดินซ้ำ
        for j in range(len(constructed_path) - 1):
            graph[constructed_path[j]] = [(node, weight) for node, weight in graph[constructed_path[j]] if node != constructed_path[j + 1]]

# กำหนดให้ตัวแปร result เก็บค่าผลลัพที่ได้ในรูปแบบ String และกำหนดให้ตัวแปร path เก็บค่าเส้นทางที่เราต้องการในรูปแบบ String
    result = ' '.join(map(str, constructed_paths[1::]))
    path = ' '.join(map(str, start[1:-1]))

    return [path, start[0], start[-1], result, total_distances]


# กำหนดให้ A=1, B=2, C=3, D=4, E=5, F=6 และ G=7 / S=รถ, X=เรือ

csv_filename = "output_file.csv"

with open(csv_filename, mode='w', newline='') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(["path", "start", "end", "shortestPath", "distance"])

    for path in final:
        graph = {
          'S': [('A', 70), ('B', 110), ('C', 240), ('D', 270), ('E', 290), ('F', 300), ('G', 350), ('X', 350)],
          'A': [('S', 70), ('B', 40), ('C', 170), ('D', 210), ('E', 270), ('F', 290), ('G', 290), ('X', 300)],
          'B': [('S', 110), ('A', 40), ('C', 140), ('D', 180), ('E', 230), ('F', 280), ('G', 260), ('X', 260)],
          'C': [('S', 240), ('A', 170), ('B', 140), ('D', 88), ('E', 140), ('F', 200), ('G', 170), ('X', 180)],
          'D': [('S', 270), ('A', 210), ('B', 180), ('C', 88), ('E', 100), ('F', 160), ('G', 130), ('X', 140)],
          'E': [('S', 290), ('A', 270), ('B', 230), ('C', 140), ('D', 100), ('F', 44), ('G', 98), ('X', 110)],
          'F': [('S', 300), ('A', 290), ('B', 280), ('C', 200), ('D', 160), ('E', 44), ('G', 55), ('X', 64)],
          'G': [('S', 350), ('A', 290), ('B', 260), ('C', 170), ('D', 130), ('E', 98), ('F', 55), ('X', 35)],
          'X': [('S', 350), ('A', 300), ('B', 260), ('C', 180), ('D', 140), ('E', 110), ('F', 64), ('G', 35)]
        }
        # dijkstra(graph, path) คือการเรียกใช้งาน funch AI
        row_data = dijkstra(graph, path)
        print(row_data)
        csv_writer.writerow(row_data)

print(f'The data has been written to {csv_filename}.')

```
